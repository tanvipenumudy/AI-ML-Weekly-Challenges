```python
df['Salary'].replace(np.NaN, df['Salary'].mean(),inplace=True)
df['Age'].replace(np.NaN, df['Age'].mean(),inplace=True)
#df.head(10)
df
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
L1=LabelEncoder()
df['Country']=L1.fit_transform(df['Country'])
df['Purchased']= L1.fit_transform(df['Purchased'])
df
```

```python
D=pd.get_dummies(df.Country)
df=pd.concat([df,D], axis=1)
df
```

```python
df.drop('Country',axis=1,inplace=True)
df
```

```python
AS=pd.DataFrame(columns=['Age','Salary'])
AS['Age']=df['Age']
AS['Salary']=df['Salary']
AS
```

```python
min_max_scaler=preprocessing.MinMaxScaler()
n=min_max_scaler.fit_transform(AS.values)
AS=pd.DataFrame(n)
AS
```

```python
df['Salary']=AS[1]
df['Age']=AS[0]
df
```
